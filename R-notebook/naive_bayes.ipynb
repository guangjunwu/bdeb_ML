```r
# training_data <- naiveBayes(x,y)
# predict(training_data, x)
data("iris")
print("=== dim ===")
dim(iris)
print("=== structure ===")
str(iris)
print("=== attributes ===")
attributes(iris)
print("=== table ===")
table(iris$Species)
```

```r
table(iris$Species)
table(iris$Sepal.Length)
```

```r
# installer package pour le naive bayes dans R-studio
# install.packages('e1071')
library(e1071)
```

```r
# training
# naiveBayes(x, y)
classifier <- naiveBayes(iris[,1:4], iris[,5]) # P(B|A)
```

```r
summary(classifier)
```

```r
classifier
```

```r
#predication
iris_prediction <- predict(classifier, iris[, -5])
iris_prediction
```

```r
# Matrice de confusion
table(iris_prediction, iris[,5], dnn=list('predicted', 'actual'))
```

```r
# calcul des distributions a priori des calsses
classifier$apriori/sum(classifier$apriori)
classifier$apriori
```
